<div class="alert alert-block alert-info">
<b>Pergunta de Neg√≥cio 5:</b> Qual segmento teve o maior total de vendas? 
<p>Demonstre o resultado atrav√©s de um gr√°fico de pizza.</p>
</div>

```python
# Agrupado por 'Segmento' para obter o total de vendas a partir desta categoria
df_dsa_prdct5 = df_dsa.groupby('Segmento')['Valor_Venda'].sum().reset_index().sort_values(by = 'Valor_Venda', ascending = False)
print(df_dsa_prdct5)
```

```python
# √â necess√°rio converter os dados para obten√ß√£o deles em valores absolutos
def convert_to_absolute(values): 
    def my_format(pct): 
        total = sum(values) 
        val = int(round(pct * total / 100.0))
        return ' $ {v:d}'.format(v = val)
    return my_format
print(convert_to_absolute)
```

```python
# Plot

# Tamanho da figura
plt.figure(figsize = (18, 6))

# Gr√°fico de pizza
plt.pie(df_dsa_prdct5['Valor_Venda'], 
        labels = df_dsa_prdct5['Segmento'],
        autopct = convert_to_absolute(df_dsa_prdct5['Valor_Venda']),
        startangle = 90)

# Limpa o c√≠rculo central
centre_circle = plt.Circle((0, 0), 0.82, fc = 'white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Labels e anota√ß√µes
plt.annotate(text = 'Total de Vendas: ' + '$ ' + str(int(sum(df_dsa_prdct5['Valor_Venda']))), xy = (-0.25, 0))
plt.title('Total de Vendas Por Segmento')
plt.show()
```

<b>Pergunta de Neg√≥cio 7:</b> Os  gestores  da  empresa  est√£o  considerando conceder  diferentes  faixas  de  descontos  e gostariam de fazer uma simula√ß√£o com base na regra abaixo: 
</b> 
<p><p>Se o Valor_Venda for maior que 1000 recebe 15% de desconto.</p></p> 
<p>Se o Valor_Venda for menor que 1000 recebe 10% de desconto.</p>
<p>Quantas vendas receberiam 15% de desconto?</p>
</div>

```python
# Criando uma coluna de acordo com a regra definida na quest√£o
df_dsa['Desconto'] = np.where(df_dsa['Valor_Venda'] > 1000, 0.15, 0.1)
```

```python
df_dsa.head(15)
```

```python
# Mostra o total por cada valor da coluna 'Desconto'
df_dsa['Desconto'].value_counts()
```

```python
print('As vendas totais com desconto de 15% seria o montante de 457.')
```

<div class="alert alert-block alert-info">
<b>Pergunta de Neg√≥cio 8:</b> Considere  que  a  Empresa  decida  conceder  o  desconto  de  15%  do  item  anterior.  Qual seria a m√©dia do valor de venda antes e depois do desconto?
</div>

```python
# Criada coluna de c√°lculo do valor da venda menos o desconto
df_dsa['Valor_Venda_Desconto'] = df_dsa['Valor_Venda'] - (df_dsa['Valor_Venda'] * df_dsa['Desconto'])
```

```python
df_dsa.head()
```

```python
# Filtro das vendas antes do desconto de 15%
df_dsa_prdct7_before_sale = df_dsa.loc[df_dsa['Desconto'] == 0.15, 'Valor_Venda']
```

```python
df_dsa_prdct7_after_sale = df_dsa.loc[df_dsa['Desconto'] == 0.15, 'Valor_Venda_Desconto']
```

```python
# C√°lculo da m√©dia das vendas antes do desconto de 15%
avg_sales_before = df_dsa_prdct7_before_sale.mean()
```

```python
# C√°lculo da m√©dia das vendas depois do desconto de 15%
avg_sales_after = df_dsa_prdct7_after_sale.mean()
```

```python
print('M√©dia das vendas antes do desconto de 15%:', round(avg_sales_before, 2))
```

```python
print('M√©dia das vendas depois do desconto de 15%:', round(avg_sales_after, 2))
```

<b>Pergunta de Neg√≥cio 9:</b> Qual a m√©dia de vendas por segmento, por ano e por m√™s? 
<p>Demonstre o resultado atrav√©s de gr√°fico de linha.</p>

```python
# Extra√ß√£o da coluna m√™s e grava√ß√£o em nova vari√°vel
df_dsa['Mes'] = df_dsa['Data_Pedido'].dt.month
```

```python
df_dsa.head()
```

```python
# Agrupamento por ano, m√™s e segmento e c√°lculo estat√≠stico de agrega√ß√£o
df_dsa_prdct8 = df_dsa.groupby(['Ano', 'Mes', 'Segmento'])['Valor_Venda'].agg([np.sum, np.mean, np.median])
```

```python
df_dsa_prdct8
```

```python
# Extra√ß√£o dos n√≠veis
yr = df_dsa_prdct8.index.get_level_values(0)
mnth = df_dsa_prdct8.index.get_level_values(1)
sgmnts = df_dsa_prdct8.index.get_level_values(2)
```

```python
# Plot

plt.figure(figsize = (10,6))
sns.set()
fig1 = sns.relplot(kind = 'line',
                   data = df_dsa_prdct8,
                   y = 'mean',
                   x = mnth,
                   hue = sgmnts,
                   col = yr,
                   col_wrap = 4)
plt.show()
```

<p>Demonstre tudo atrav√©s de um √∫nico gr√°fico.</p>

```python
df_dsa.head()
```

```python
# Agrupamento e soma
df_dsa_prdctfinal = df_dsa.groupby(['Categoria',
                             'SubCategoria']).sum(numeric_only = True).sort_values('Valor_Venda',
                                                                                   ascending = False).head(12)
```

```python
# Convers√£o da coluna Valor_Venda em n√∫mero int e classifica√ß√£o por categoria
df_dsa_prdctfinal = df_dsa_prdctfinal[['Valor_Venda']].astype(int).sort_values(by = 'Categoria').reset_index()
```

```python
df_dsa_prdctfinal
```

```python
# Criando outro dataframe com totais por categoria
df_dsa_prdctfinal_cat = df_dsa_prdctfinal.groupby('Categoria').sum(numeric_only = True).reset_index()
```

```python
df_dsa_prdctfinal_cat
```

```python
# Listas de cores para categorias: #845EC2, #F3C5FF, #00C9A7
color_ctg = ['#845EC2',
                    '#F3C5FF',
                    '#00C9A7']
```

```python
# Listas de cores para subcategorias

color_subctg = ['#6B4A9A',
                       '#A07BCB',
                       '#523D7E',
                       '#8D6FAF', #
                       '#DAA9E6',
                       '#F8D9FF',
                       '#CBA0E0',
                       '#E6BFEF', #
                       '#00997F',
                       '#00E3C0',
                       '#008A6E',
                       '#00B58F']
```

```python
# Plot

# Tamanho da figura
fig, ax = plt.subplots(figsize = (18,12))

# Gr√°fico das categorias
p1 = ax.pie(df_dsa_prdctfinal_cat['Valor_Venda'], 
            radius = 1,
            labels = df_dsa_prdctfinal_cat['Categoria'],
            wedgeprops = dict(edgecolor = 'white'),
            colors = color_ctg)

# Gr√°fico das subcategorias
p2 = ax.pie(df_dsa_prdctfinal['Valor_Venda'],
            radius = 0.9,
            labels = df_dsa_prdctfinal['SubCategoria'],
            autopct = convert_to_absolute(df_dsa_prdctfinal['Valor_Venda']),
            colors = color_subctg, 
            labeldistance = 0.7,
            wedgeprops = dict(edgecolor = 'white'), 
            pctdistance = 0.53,
            rotatelabels = True)

# Limpa o centro do c√≠rculo
centre_circle = plt.Circle((0, 0), 0.6, fc = 'white')

# Labels e anota√ß√µes
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.annotate(text = 'Total de Vendas: ' + '$ ' + str(int(sum(df_dsa_prdctfinal['Valor_Venda']))), xy = (-0.2, 0))
plt.title('Total de Vendas Por Categoria e Top 12 SubCategorias')
plt.show()
```
